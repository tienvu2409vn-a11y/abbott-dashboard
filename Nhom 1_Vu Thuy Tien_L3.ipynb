```python
#bài 3
#append(x) → thêm x vào cuối list.
#insert(i, x) → chèn x vào vị trí i trong list.

numbers = []# Tạo list rỗng

# Thêm số 3
numbers.append(3)
print("Sau khi thêm 3:", numbers)

# Thêm số 7
numbers.append(7)
print("Sau khi thêm 7:", numbers)

# Thêm số 5
numbers.append(5)
print("Sau khi thêm 5:", numbers)

# Chèn số 10 vào vị trí đầu tiên
numbers.insert(0, 10)
print("Sau khi chèn 10 vào đầu list:", numbers)
```

```python
# bài 7
numbers = [9, 3, 11, 5, 2]
max_num = numbers[1]        # giả sử số đầu tiên là lớn nhất
for i in numbers:
    if i > max_num:         # so sánh với max hiện tại
        max_num = i

print("Số lớn nhất:", max_num)
```
