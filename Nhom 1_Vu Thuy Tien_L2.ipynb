```python
#bài 5
x=float(input("nhập cạnh 1"))
y=float(input("nhập cạnh 2"))
z=float(input("nhập cạnh 3"))

if x + y > z and x + z > y and y + z > x:
    if x == y == z:
        print("Tam giác đều")
    elif x == y or y == z or z == x:
        print("Tam giác cân")
    elif x**2 == y**2 + z**2 or y**2 == x**2 + z**2 or z**2 == x**2 + y**2:
        print("Tam giác vuông")
    elif x == y or y == z or z == x :
        if x**2 == y**2 + z**2 or y**2 == x**2 + z**2 or z**2 == x**2 + y**2:
            print("Tam giác vuông cân")
        else:
            print("Tam giác cân")
    else:
        print("Tam giác thường")
else:
    print("Không phải tam giác")
```

```python
#bài 8
n=int(input("Nhập số lượng phần tử n:"))
tong = 0
for i in range (n):
    x = int(input(f"Nhập số thứ {i+1}: "))
    tong += x
print(f"Tổng của {n} số đã nhập là: {tong}")
```
