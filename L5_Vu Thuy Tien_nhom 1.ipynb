```python
# Bài 4: 
n = int(input("Nhập số lượng phần tử: "))

ds = []
for i in range(n):
    so = int(input("Nhập số: "))
    ds.append(so)

# Loại bỏ trùng lặp bằng cách chuyển sang set rồi chuyển lại thành list
ds_khong_trung = list(set(ds))

print("Danh sách sau khi loại bỏ trùng lặp:", ds_khong_trung)
```

```python
#bài 5

n = int(input("Nhập số lượng phần tử: "))

ds = []
for i in range(n):
    so = int(input("Nhập số: "))
    ds.append(so)

# Loại bỏ trùng lặp bằng cách chuyển sang set rồi chuyển lại thành list
ds_dao_nguoc = list(reversed(ds))

print("Danh sách sau khi loại đảo ngược:", ds_dao_nguoc)
```

```python
# Bài 7: 
n = int(input("Nhập số lượng phần tử: "))

ds = []
for i in range(n):
    so = int(input("Nhập số: "))
    ds.append(so)

tim = int(input("Nhập số cần tìm: "))
so_trung_lap = ds.count(tim)   

print(f"Số {tim} xuất hiện {so_trung_lap} lần trong danh sách.")

```

```python
# Tách số chẵn và số lẻ trong danh sách
n = int(input("Nhập số lượng phần tử: "))

ds = [] 
for i in range(n):
    so = int(input("Nhập số: "))
    ds.append(so)

chan = []
le = []
for x in ds:
    if x % 2 == 0:
        chan.append(x)
    else:
        le.append(x)

print("Danh sách số chẵn:", chan)
print("Danh sách số lẻ:", le)

```

```python
# Bài 10: 
chuoi = input("Nhập chuỗi: ")

tan_so = {}   # tạo từ điển rỗng
for ky_tu in chuoi:
    if ky_tu in tan_so:
        tan_so[ky_tu] += 1   # nếu đã có thì tăng thêm 1
    else:
        tan_so[ky_tu] = 1    # nếu chưa có thì gán bằng 1

print(tan_so)
```
